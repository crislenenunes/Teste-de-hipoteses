```python
# Exercício 1 - Teste de Hipótese para Média (Z-Test)
import math
import scipy.stats as sp

# Dados do problema
mu_0 = 500  # Média populacional
x_barra = 498  # Média amostral
sigma = 10  # Desvio padrão populacional
n = 200  # Tamanho da amostra

# Erro padrão
erro_padrao = sigma / math.sqrt(n)

# Estatística do teste (z-test)
z_test = (x_barra - mu_0) / erro_padrao

# Valor crítico para um teste bicaudal de 5% de significância
z_critico = sp.norm.ppf(1 - 0.025)

# Decisão
if abs(z_test) > z_critico:
    resultado = "Rejeitamos H0"
else:
    resultado = "Não rejeitamos H0"

# Valor de p (p-valor)
p_valor = 2 * (1 - sp.norm.cdf(abs(z_test)))

# Exibição dos resultados
print("Exercício 1:")
print(f"Estatística do teste: {z_test:.4f}")
print(f"Valor crítico: {z_critico:.4f}")
print(f"Decisão: {resultado}")
print(f"Valor de p: {p_valor:.4f}")
print("\n")

# Exercício 2 - Teste de Hipótese para Média (Z-Test)
mu_0 = 1200
x_barra = 1265
sigma = 300
n = 100

# Erro padrão
erro_padrao = sigma / math.sqrt(n)

# Estatística do teste (z-test)
z_test = (x_barra - mu_0) / erro_padrao

# Valor crítico para um teste unilateral de 5% de significância
z_critico = sp.norm.ppf(1 - 0.05)

# Decisão
if z_test > z_critico:
    resultado = "Rejeitamos H0"
else:
    resultado = "Não rejeitamos H0"

# Valor de p (p-valor)
p_valor = 1 - sp.norm.cdf(z_test)

# Exibição dos resultados
print("Exercício 2:")
print(f"Estatística do teste: {z_test:.4f}")
print(f"Valor crítico: {z_critico:.4f}")
print(f"Decisão: {resultado}")
print(f"Valor de p: {p_valor:.4f}")
print("\n")

# Exercício 3 - Teste de Hipótese para Média (Z-Test)
mu_0 = 42
x_barra = 34.9
s = 3.96
n = 100

# Erro padrão
erro_padrao = s / math.sqrt(n)

# Estatística do teste (z-test)
z_test = (x_barra - mu_0) / erro_padrao

# Valor crítico para um teste unilateral de 5% de significância
z_critico = sp.norm.ppf(0.05)

# Decisão
if z_test < z_critico:
    resultado = "Rejeitamos H0"
else:
    resultado = "Não rejeitamos H0"

# Valor de p (p-valor)
p_valor = sp.norm.cdf(z_test)

# Exibição dos resultados
print("Exercício 3:")
print(f"Estatística do teste: {z_test:.4f}")
print(f"Valor crítico: {z_critico:.4f}")
print(f"Decisão: {resultado}")
print(f"Valor de p: {p_valor:.4f}")
print("\n")

# Exercício 4 - Teste de Hipótese para Média (Z-Test)
mu_0 = 57
x_barra = 56.91
s = 0.24
n = 100

# Erro padrão
erro_padrao = s / math.sqrt(n)

# Estatística do teste (z-test)
z_test = (x_barra - mu_0) / erro_padrao

# Valor crítico para um teste bicaudal de 5% de significância
z_critico = sp.norm.ppf(1 - 0.025)

# Decisão
if abs(z_test) > z_critico:
    resultado = "Rejeitamos H0"
else:
    resultado = "Não rejeitamos H0"

# Valor de p (p-valor)
p_valor = 2 * (1 - sp.norm.cdf(abs(z_test)))

# Exibição dos resultados
print("Exercício 4:")
print(f"Estatística do teste: {z_test:.4f}")
print(f"Valor crítico: {z_critico:.4f}")
print(f"Decisão: {resultado}")
print(f"Valor de p: {p_valor:.4f}")
print("\n")

# Exercício 5 - Teste de Hipótese para Proporções (Z-Test)
p_0 = 0.32
sucessos = 145
n = 500

# Proporção amostral
p_hat = sucessos / n

# Erro padrão
erro_padrao = math.sqrt((p_0 * (1 - p_0)) / n)

# Estatística do teste (z-test)
z_test = (p_hat - p_0) / erro_padrao

# Valor crítico para um teste unilateral de 2% de significância
z_critico = sp.norm.ppf(0.02)

# Decisão
if z_test < z_critico:
    resultado = "Rejeitamos H0"
else:
    resultado = "Não rejeitamos H0"

# Valor de p (p-valor)
p_valor = sp.norm.cdf(z_test)

# Exibição dos resultados
print("Exercício 5:")
print(f"Estatística do teste: {z_test:.4f}")
print(f"Valor crítico: {z_critico:.4f}")
print(f"Decisão: {resultado}")
print(f"Valor de p: {p_valor:.4f}")
print("\n")

# Exercício 6 - Teste de Hipótese para Proporções (Z-Test)
p_0 = 0.06
substituicoes = 12
n = 400

# Proporção amostral
p_hat = substituicoes / n

# Erro padrão
erro_padrao = math.sqrt((p_0 * (1 - p_0)) / n)

# Estatística do teste (z-test)
z_test = (p_hat - p_0) / erro_padrao

# Valor crítico para um teste unilateral de 3% de significância
z_critico = sp.norm.ppf(0.03)

# Decisão
if z_test < z_critico:
    resultado = "Rejeitamos H0"
else:
    resultado = "Não rejeitamos H0"

# Valor de p (p-valor)
p_valor = sp.norm.cdf(z_test)

# Exibição dos resultados
print("Exercício 6:")
print(f"Estatística do teste: {z_test:.4f}")
print(f"Valor crítico: {z_critico:.4f}")
print(f"Decisão: {resultado}")
print(f"Valor de p: {p_valor:.4f}")
print("\n")

# Exercício 7 - Teste de Hipótese para Proporções (Z-Test)
p_0 = 0.06
substituicoes = 12
n = 400

# Proporção amostral
p_hat = substituicoes / n

# Erro padrão
erro_padrao = math.sqrt((p_0 * (1 - p_0)) / n)

# Estatística do teste (z-test)
z_test = (p_hat - p_0) / erro_padrao

# Valor crítico para um teste unilateral de 3% de significância
z_critico = sp.norm.ppf(0.03)

# Decisão
if z_test < z_critico:
    resultado = "Rejeitamos H0"
else:
    resultado = "Não rejeitamos H0"

# Valor de p (p-valor)
p_valor = sp.norm.cdf(z_test)

# Exibição dos resultados
print("Exercício 7:")
print(f"Estatística do teste: {z_test:.4f}")
print(f"Valor crítico: {z_critico:.4f}")
print(f"Decisão: {resultado}")
print(f"Valor de p: {p_valor:.4f}")
```
